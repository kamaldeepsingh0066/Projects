
# Global Economy Indicators - Data Preparation

This notebook performs data preparation on the Global Economy Indicators dataset. The steps include:

1. Loading the dataset
2. Cleaning column names
3. Filtering data for the last 5 years (2017-2021)
4. Aggregating the data by country
5. Selecting relevant variables
6. Normalizing industry sector data as % of GDP
7. Exporting the final prepared dataset



## Step 1: Load the dataset

We start by importing the required libraries and loading the CSV file.


```python

import pandas as pd

# Load the dataset (adjust the file path if needed)
file_path = 'Global Economy Indicators.csv'
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```


## Step 2: Clean column names

Remove any leading and trailing spaces from column headers for easier processing.


```python

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Verify column names after stripping spaces
df.columns.tolist()
```


## Step 3: Filter data for the last 5 years (2017-2021)

We want to calculate the mean values over the most recent 5 years for each country.


```python

# Filter the dataset
recent_df = df[df['Year'] >= 2017]

# Display filtered data
recent_df.head()
```


## Step 4: Aggregate data by country

Compute the mean for each numeric column grouped by country.


```python

# Group by Country and calculate mean
recent_grouped = recent_df.groupby('Country').mean(numeric_only=True).reset_index()

# Display the aggregated data
recent_grouped.head()
```


## Step 5: Select relevant columns for analysis

We select economic indicators and industry sector columns for further analysis.


```python

columns_of_interest = [
    'Country',
    'Population',
    'Per capita GNI',
    'Gross National Income(GNI) in USD',
    'Gross Domestic Product (GDP)',
    'Agriculture, hunting, forestry, fishing (ISIC A-B)',
    'Manufacturing (ISIC D)',
    'Mining, Manufacturing, Utilities (ISIC C-E)',
    'Other Activities (ISIC J-P)',
    'Transport, storage and communication (ISIC I)',
    'Wholesale, retail trade, restaurants and hotels (ISIC G-H)',
    'Construction (ISIC F)'
]

cleaned_df = recent_grouped[columns_of_interest].copy()
cleaned_df.head()
```


## Step 6: Normalize industry sector data as % of GDP

We convert absolute sector values into percentages of GDP to make them comparable across countries.


```python

industry_columns = [
    'Agriculture, hunting, forestry, fishing (ISIC A-B)',
    'Manufacturing (ISIC D)',
    'Mining, Manufacturing, Utilities (ISIC C-E)',
    'Other Activities (ISIC J-P)',
    'Transport, storage and communication (ISIC I)',
    'Wholesale, retail trade, restaurants and hotels (ISIC G-H)',
    'Construction (ISIC F)'
]

for col in industry_columns:
    cleaned_df[col + ' (%)'] = (cleaned_df[col] / cleaned_df['Gross Domestic Product (GDP)']) * 100

# Drop absolute value columns and keep percentages only
cleaned_df = cleaned_df.drop(columns=industry_columns)

# Display cleaned data
cleaned_df.head()
```


## Step 7: Export prepared dataset

Save the cleaned and prepared data to a CSV file for further analysis.


```python

output_file_path = 'Prepared_Global_Economy_Indicators.csv'
cleaned_df.to_csv(output_file_path, index=False)

print("Data preparation complete. File exported to:", output_file_path)
```
